# 02 — Model Training & Evaluation

Train multiple classifiers, compare them, pick the best, evaluate on the held-out test set, and persist the final model as `model/model.pkl`.

```python
import pandas as pd, numpy as np, joblib, json, os
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report)
```

## 1. Load Cleaned Splits

```python
train = pd.read_csv('../dataset/train_clean.csv')
test  = pd.read_csv('../dataset/test_clean.csv')
X_train, y_train = train.drop('target',axis=1), train['target']
X_test,  y_test  = test.drop('target',axis=1),  test['target']
print(X_train.shape, X_test.shape)
```

## 2. Candidate Models
We try Logistic Regression, KNN, Decision Tree, Random Forest, and SVM.

```python
models = {
    'LogReg'      : LogisticRegression(max_iter=1000),
    'KNN'         : KNeighborsClassifier(),
    'DecisionTree': DecisionTreeClassifier(random_state=42),
    'RandomForest': RandomForestClassifier(n_estimators=200, random_state=42),
    'SVM'         : SVC(probability=True, random_state=42),
}
```

## 3. 5-Fold Cross-Validation Comparison

```python
rows=[]
for name, m in models.items():
    scores = cross_val_score(m, X_train, y_train, cv=5, scoring='accuracy')
    rows.append((name, scores.mean(), scores.std()))
cv_df = pd.DataFrame(rows, columns=['model','cv_mean','cv_std']).sort_values('cv_mean', ascending=False)
cv_df
```

## 4. Hyperparameter Tuning (Random Forest)

```python
grid = GridSearchCV(RandomForestClassifier(random_state=42),
    {'n_estimators':[100,200,300], 'max_depth':[None,5,10]},
    cv=5, scoring='accuracy', n_jobs=1)
grid.fit(X_train, y_train)
print('Best params:', grid.best_params_, 'CV acc:', grid.best_score_)
best = grid.best_estimator_
```

## 5. Final Evaluation on Test Set

```python
pred = best.predict(X_test)
print(f'Accuracy : {accuracy_score(y_test,pred):.4f}')
print(f'Precision: {precision_score(y_test,pred):.4f}')
print(f'Recall   : {recall_score(y_test,pred):.4f}')
print(f'F1-score : {f1_score(y_test,pred):.4f}')
print('Confusion Matrix:\n', confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

## 6. Save Final Model

```python
os.makedirs('../model', exist_ok=True)
joblib.dump(best, '../model/model.pkl')
with open('../model/features.json','w') as f:
    json.dump(list(X_train.columns), f)
print('✅ Saved model/model.pkl')
```

## 7. Sanity-Check the Pickle

```python
loaded = joblib.load('../model/model.pkl')
sample = X_test.iloc[[0]]
print('Prediction:', loaded.predict(sample)[0], '| True:', y_test.iloc[0])
```

**Conclusion:** Random Forest gives the best balance of accuracy/recall and is suitable for a screening tool — recall on the positive (disease) class is the priority because false negatives carry the highest clinical cost.
