# Model Comparison
Compare Random Forest, Logistic Regression and SVM on the three datasets.


```python
import json, os
from pathlib import Path
metrics = json.loads(Path('../saved_models/metrics.json').read_text())
metrics
```

```python
import pandas as pd
rows = []
for d, info in metrics.items():
    for m, r in info['results'].items():
        rows.append({'disease': d, 'model': m, 'accuracy': r['accuracy'], 'cv': r['cv_accuracy']})
pd.DataFrame(rows).pivot(index='model', columns='disease', values='accuracy')
```
